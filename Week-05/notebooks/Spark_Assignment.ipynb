# Assignment Questions (Q1–Q15)


## Q1. What are the key limitations of traditional MapReduce?

### Answer

Traditional MapReduce has several limitations that make Apache Spark a preferred framework for modern big data processing.

- Stores intermediate results on disk after every stage.
- High disk I/O results in slower execution.
- Not suitable for iterative Machine Learning algorithms.
- Complex programming model using separate Map and Reduce phases.
- High latency for interactive analytics.
- Limited support for real-time data processing.

### Why Spark?

- In-memory computation
- Faster processing
- DAG execution engine
- Rich DataFrame API
- Supports SQL, Streaming, MLlib, and Graph Processing

## Q2. Spark In-Memory Computing

Apache Spark stores intermediate data in RAM instead of writing it to disk after every computation.

Advantages include:

- Faster execution
- Reduced disk I/O
- Better CPU utilization
- Efficient iterative Machine Learning
- Improved performance for big data analytics

Compared to MapReduce, Spark significantly reduces execution time by caching intermediate datasets in memory.

```python
print("Before:", df.count())

df_q3 = df.dropDuplicates(["product_id"])

print("After:", df_q3.count())
```

```python
from pyspark.sql.functions import avg

result = (
    df
    .filter(col("rating") >= 4)
    .groupBy("category")
    .agg(avg("final_price").alias("Average_Final_Price"))
)

result.show(truncate=False)
```

## Q5. Difference between `.na.drop()` and `.na.fill()`

### `.na.drop()`

Removes rows containing null values.

### `.na.fill()`

Replaces null values with specified values instead of removing records.

Using `.na.fill()` helps preserve data while handling missing information.

```python
df_q5 = df.na.fill({
    "seller_name": "Unknown"
})

df_q5.select("seller_name").show(10, truncate=False)
```

```python
from pyspark.sql.functions import count

category_count = (
    df
    .groupBy("category")
    .agg(count("*").alias("Total_Products"))
    .filter(col("Total_Products") > 10)
)

category_count.show(truncate=False)
```

## Q7. How does DataFrame immutability affect data cleaning?

Spark DataFrames are immutable.

Operations like:

- Dropping columns
- Renaming columns
- Filtering rows
- Filling null values

do not modify the original DataFrame.

Instead, Spark creates a new transformed DataFrame.

This improves fault tolerance and enables Spark's optimization engine.

```python
premium_products = df.filter(
    (col("rating") >= 4) &
    (col("discount") >= 20)
)

premium_products.select(
    "product_id",
    "title",
    "rating",
    "discount"
).show()
```

## Q9. Why should null values be handled before aggregation?

Handling null values before performing aggregation ensures accurate analytical results.

Benefits include:

- Accurate averages
- Correct sum calculations
- Reliable statistical analysis
- Better data quality
- Improved reporting

```python
from pyspark.sql.types import DoubleType

df_q10 = (
    df
    .withColumn(
        "price",
        col("final_price").cast(DoubleType())
    )
)

df_q10.printSchema()
```

## Q11. What is Shuffle?

Shuffle is the process of redistributing data across partitions during operations such as:

- groupBy()
- join()
- distinct()

Shuffle is considered a wide transformation because data moves between different partitions and worker nodes.

Although necessary, shuffle is computationally expensive due to network communication and disk usage.

```python
df_q12 = df.filter(
    col("seller_name").isNotNull() &
    (col("title") != "")
)

df_q12.show(10, truncate=False)
```

```python
from pyspark.sql.functions import min, max, mean

df.agg(
    min("final_price").alias("Minimum Price"),
    max("final_price").alias("Maximum Price"),
    mean("final_price").alias("Average Price")
).show()
```

## Q14. Risk of using inferSchema=True

When datasets contain inconsistent values or mixed data types, Spark may infer incorrect column types.

Possible issues include:

- Incorrect numeric types
- Date columns read as strings
- Null values created due to parsing failures
- Schema inconsistencies

For production systems, defining an explicit schema is often recommended.

```python
from pyspark.sql.functions import sum

final_pipeline = (
    df
    .dropDuplicates(["product_id"])
    .na.fill({
        "final_price": 0,
        "seller_name": "Unknown"
    })
    .groupBy("category")
    .agg(
        sum("final_price").alias("Total Revenue")
    )
    .orderBy(desc("Total Revenue"))
)

final_pipeline.show(truncate=False)
```
